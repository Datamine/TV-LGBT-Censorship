```python
from ast import literal_eval
import get_annual_totals
from matplotlib import pyplot
from os import listdir
from ast import literal_eval
%matplotlib inline
```

```python
year_counts = get_annual_totals.all_years("processed_year_wordcounts/")
# 2015 has barely any words so far, so we removed it from consideration for this part.
del year_counts['2015']
```

```python
keys = sorted(year_counts.keys())
values = [year_counts[k] for k in keys]
pyplot.figure(figsize=(12,8))
pyplot.scatter(keys,values)
pyplot.plot(keys,values)
pyplot.ylim(0,140000)
pyplot.xlim(1950,2020)
pyplot.xlabel("Year")
pyplot.ylabel("Word Frequency Counts, Stopwords Removed")
pyplot.show()
```

```python
path = "processed_year_wordcounts/"
directory = sorted(listdir(path))

def get_year_counts(word):
    # proportional
    year_counts = {}
    for year in directory:
        with open(path + year, 'r') as f:
            dic = literal_eval(f.read().split('\n')[0])
            total = float(sum(dic.values()))
            if word in dic:
                year_counts[year] = dic[word]/total
            else:
                year_counts[year] = 0
    del year_counts['2015']
    keys = sorted(year_counts.keys())
    values = [year_counts[k] for k in keys]
    return keys, values
```

```python
pyplot.figure(figsize=(12,8))

words = ["marriage","marry","wed","honeymoon"]

for w in words:
    x, y = get_year_counts(w)
    if w=='marri':
        print x,y
    #print w, y, "\n"
    #pyplot.scatter(x,y,label=w)
    pyplot.plot(x,y,marker="o",label=w)

pyplot.legend(words)
pyplot.xlim(1970,2020)
#pyplot.ylim(0,0.002)
pyplot.xlabel("Year")
pyplot.ylabel("Word Frequency Counts, Stopwords Removed")
pyplot.show()
```
